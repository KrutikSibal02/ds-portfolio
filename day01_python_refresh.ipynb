```python
# Write a function that takes a salary and returns
# whether it's above or below your 5 LPA target
def salary_check(lpa):
    if lpa >= 5:
        return f"{lpa} LPA — Target met!"
    else:
        return f"{lpa} LPA — Below target, keep searching."

print(salary_check(4))
print(salary_check(6))
print(salary_check(5))
```

```python
def Master_level(lufs):
  if lufs <= -14:
    return f"{lufs} Optimal Master for Release"
  elif lufs<= -10:
    return f"{lufs} Loud Master"
  else :
    return f"{lufs} Too loud Master"

print(Master_level(-10))
print(Master_level(-6))
print(Master_level(-9))
print(Master_level(-12))
print(Master_level(-14))
```

```python
def plays(streams):
  if streams >= 1000000:
    return f"{streams} views - viral"
  elif streams >= 50000:
    return f"{streams} views - Acceptable"
  else :
    return f"{streams} views - Needs Improvement"

print(plays(19000000))
print(plays(6900000))
print(plays(54000))
print(plays(14000))
print(plays(170))
```
